```python
def binary_to_decimal(binary_array):
    decimal_number = 0
    binary_array = binary_array[::-1]  # 反转数组以便从低位到高位计算

    for i in range(len(binary_array)):
        decimal_number += binary_array[i] * (2 ** i)

    return decimal_number
code = []
for i in range(400,800):
    data = test[i,6:11]
    binary_array = np.array(data, dtype=np.uint8)
    
    decimal_number = binary_to_decimal(binary_array)
    print(decimal_number)
    code.append(decimal_number)
plt.plot(code)
plt.plot(label[400:800])
```
